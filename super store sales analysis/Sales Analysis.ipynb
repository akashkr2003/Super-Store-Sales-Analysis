```python
# method that counts the unique occurance of values in the series 
# .reset_index() this gives the index 0,1,2 
# .value_counts() this gives the count 
customer = df["Segment"].value_counts().reset_index()
# incase of rename we will write code
# customer  = customer.rename(columns = {"Segment":"new name" ,"count":"new"})
print(customer)
```

```python
# Customer and sales 
```

```python
sales = df.groupby("Segment")["Sales"].sum().reset_index()
sales = sales.rename(columns = {"Segment":"Customer Type","Sales":"Total Sales"})

print(sales)                
```

```python
# plotting the second pie chart  
c = {"r","b","g"}
plt.pie(sales["Total Sales"],labels = sales["Customer Type"],autopct = "%0.1f%%",shadow = True,textprops={"fontsize": 10},colors = c)
plt.title("Sales Per Customer Category")
plt.legend()
plt.show()
```

```python
# Bar plot representation of Sales Per Customer Category
c = {"r","b","g"}
plt.bar(sales["Customer Type"],sales["Total Sales"],color = c)

plt.title("Sales Per Customer Category")
plt.xlabel("Customer Type")
plt.ylabel("Total Sales")
plt.show()
```

```python
# Customer loyalty (repeat purchase )
```

```python
# group data according to : customer id , customer name , segment and calculate customer freqi of their order 
```

```python
corder = df.groupby(["Customer ID","Customer Name","Segment"])["Order ID"].count().reset_index()
#renaming 
corder.rename(columns = {"Order ID":"Total Orders"},inplace = True)
#identifying repeat customer 
repeat_customer = corder[corder["Total Orders"]>=1]

#sort repeated orders in descending order 

sort_repeat_customer = repeat_customer.sort_values(by = "Total Orders",ascending = False)

# wht drop method will do is it will drop the earlier reset index (787,606,482,etc)
print(sort_repeat_customer.head(10).reset_index(drop = True)
```

```python
# ranking best customer on the basis of sales 
```

```python
csales = df.groupby(["Customer ID","Customer Name","Segment"])["Sales"].sum().reset_index()

#sorting in descending order
sort_highest_sales = csales.sort_values(by = "Sales",ascending = False)
print(sort_highest_sales.head(10).reset_index(drop = True))
```

```python
# Sales trend analysis
```

```python
# convert order date into date time format
df["Order Date"] = pd.to_datetime(df["Order Date"],dayfirst = True)

yearly = df.groupby(df["Order Date"].dt.year)["Sales"].sum()

# setting index and renaming the columns 

yearly = yearly.reset_index()
yearly = yearly.rename(columns={"Order Date":"Year","Sales":"Total Sales"})

print(yearly)
```

```python
# plotting a bar graph 

plt.bar(yearly["Year"],yearly["Total Sales"],color = "r",edgecolor = "b",width = 0.4)
plt.title("Yearly Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
# plotting a line graph 

plt.plot(yearly["Year"],yearly["Total Sales"],color = "r",marker = "o",linestyle = "--")
plt.title("Yearly Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.show()
```

```python
# OUARTELY SALES 
```

```python
# convert order date into date time format
df["Order Date"] = pd.to_datetime(df["Order Date"],dayfirst = True)

# filter data according to year 

yearly = df[df["Order Date"].dt.year == 2018]

# calculate quarterly sales for year 2018

quarterly = yearly.resample("QE", on= "Order Date")["Sales"].sum()
quarterly = quarterly.reset_index()
quarterly = quarterly.rename(columns={"Order Date":"Quarter","Sales":"Total Sales"})

print("This is the quater sales for year 2018")
print(quarterly)
```

```python
# plotting a line graph 

plt.plot(quarterly["Quarter"],quarterly["Total Sales"],color = "r",marker = "o",linestyle = "--")
plt.title("quarterly Sales")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.xticks(rotation = 65)
plt.show()
```

```python
## Montly sales 
```

```python
# convert order date into date time format
df["Order Date"] = pd.to_datetime(df["Order Date"],dayfirst = True)
yearly = df[df["Order Date"].dt.year == 2018]

# calculate monthly sales for year 2018

monthly = yearly.resample("ME", on= "Order Date")["Sales"].sum()
monthly = monthly.reset_index()
monthly = monthly.rename(columns={"Order Date":"Monthly","Sales":"Total Sales"})

print("This is the monthly sales for year 2018")
print(monthly)
```

```python
# plotting a line graph 

plt.plot(monthly["Monthly"],monthly["Total Sales"],color = "r",marker = "o",linestyle = "--")
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales")
# plt.xticks(rotation = 65)
plt.show()
```
